```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("spotify_songs_dataset.csv")

# Find invalid duration
invalid_count = df['duration'].isna().sum() + (df['duration'] == "").sum()
print(f"Invalid Duration : {invalid_count}\n")

# Remove invalid duration
df = df[~(df['duration'].isna() | (df['duration'] == ""))]

# Exploratory Data Analysis (EDA) - Bar Chart Comparison of Song Duration Category
# Define category
def categorize_duration(duration):
    if duration < 120:  
        return 'Short'
    elif 120 <= duration <= 240: 
        return 'Medium'
    elif 240 < duration <= 360: 
        return 'Long'
    else: 
        return 'Very Long'
df['Category'] = df['duration'].apply(categorize_duration)

# Dataset
df['Category'] = df['duration'].apply(categorize_duration)
comparison = df['Category'].value_counts()
    
# Bar Chart
plt.figure(figsize=(10, 6))
ax = comparison.plot(kind='bar')
# Add label
for i, value in enumerate(comparison):
    plt.text(i, value + 1, str(value), ha='center', va='bottom')
plt.title("Comparison of Song Duration Category")
plt.xlabel("Category")
plt.ylabel("Total Song")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import statistics
import numpy as np
import matplotlib.pyplot as plt

# Declaration
# Exploratory Data Analysis (EDA) - Most Released Month
df = pd.read_csv("spotify_songs_dataset.csv")

# Convert Date to Month
df['Date'] = pd.to_datetime(df['release_date'], format='%Y-%m-%d', errors='coerce')
df['Month'] = df['Date'].dt.month_name()

# Dataset
top_context = df['Month'].value_counts().head(7)
    
# Bar Chart
plt.figure(figsize=(10, 6))
ax = top_context.plot(kind='bar')
# Add label
for i, value in enumerate(top_context):
    plt.text(i, value + 1, str(value), ha='center', va='bottom')
plt.title("Most Release Month")
plt.xlabel("Month")
plt.ylabel("Number of Released Song")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import statistics
import numpy as np
import matplotlib.pyplot as plt

# Declaration
# Exploratory Data Analysis (EDA) - Most Total Released Song in A Year
df = pd.read_csv("spotify_songs_dataset.csv")

# Convert Date to Month
df['Year'] = df['release_date'].apply(lambda x: x.split('-')[0].strip() if isinstance(x, str) else x)
top_context = df['Year'].value_counts().head(7)
    
# Bar Chart
plt.figure(figsize=(10, 6))
ax = top_context.plot(kind='bar')
# Add label
for i, value in enumerate(top_context):
    plt.text(i, value + 1, str(value), ha='center', va='bottom')
plt.title("Most Total Released Song in A Year")
plt.xlabel("Year")
plt.ylabel("Number of Released Song")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import re

df = pd.read_csv("spotify_songs_dataset.csv")

# Exploratory Data Analysis (EDA) - Bar Chart Most Used Word in Context for Nonconditional Data
target_col = ['song_title','album']

for dt in target_col:
    # Lower Letter and Remove Symbol
    df[dt] = df[dt].str.lower()
    df[dt] = df[dt].apply(lambda x: re.findall(r'\b\w+\b', str(x)))

    # Split Word
    words = [word for sublist in df[dt] for word in sublist]

    # Dataset
    top_context = pd.Series(words).value_counts().head(5)

    # Bar Chart
    plt.figure(figsize=(10, 6))
    ax = top_context.plot(kind='bar')
    # Add label
    for i, value in enumerate(top_context):
        plt.text(i, value + 1, str(value), ha='center', va='bottom')
    plt.title(f"Most Words in {dt}")
    plt.xlabel("Word")
    plt.ylabel("Number of Words")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
